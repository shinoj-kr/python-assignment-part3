## Task 2 — API Integration

## Step 1 — Fetch Products

This step sends a request to the API and retrieves 20 products. The response is converted into JSON format and stored for further use.


```python
# Import requests library
import requests

# Step 1: Fetch 20 products
url = "https://dummyjson.com/products?limit=20"

response = requests.get(url)  # make GET request
data = response.json()        # convert to JSON

products = data["products"]   # extract product list

print("Fetched", len(products), "products successfully.")
```

## Step 1 contd — Display Products

This step prints the fetched product data in a simple table format showing ID, title, category, price, and rating.


```python
# Step 1: Print formatted table
print("\nID | Title | Category | Price | Rating")
print("-" * 60)

for p in products:
    print(f"{p['id']} | {p['title']} | {p['category']} | ${p['price']} | {p['rating']}")
```

## Step 2 — Filter and Sort

This step filters products with rating greater than or equal to 4.5 and sorts them by price in descending order.


```python
# Step 2: Filter products with rating >= 4.5
filtered = [p for p in products if p["rating"] >= 4.5]

# Sort by price (descending)
filtered_sorted = sorted(filtered, key=lambda x: x["price"], reverse=True)

print("\nFiltered & Sorted Products:")

for p in filtered_sorted:
    print(f"{p['title']} | Price: ${p['price']} | Rating: {p['rating']}")
```

## Step 3 — Search by Category

This step fetches products from the laptops category and displays their name and price.


```python
# Step 3: Fetch laptops category
url = "https://dummyjson.com/products/category/laptops"

response = requests.get(url)
data = response.json()

laptops = data["products"]

print("\nLaptops:")

for p in laptops:
    print(f"{p['title']} | Price: ${p['price']}")
```

## Task 3 — Exception Handling

## Part A — Guarded Calculator:
## Step 1 — Safe Divide

This step creates a function to safely divide two numbers. It handles errors like division by zero and invalid input types without crashing the program.


```python
# Step 1: Safe divide function
def safe_divide(a, b):
    try:
        return a / b  # normal division
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


# Testing the function
print(safe_divide(10, 2))     # valid
print(safe_divide(10, 0))     # zero division
print(safe_divide("ten", 2))  # type error
```

## Part B — Guarded File Reader:
## Step 2 — Safe File Reader

This step reads a file safely. If the file is not found, it shows an error message and ensures the program continues without crashing.


```python
# Step 2: Safe file reader
def read_file_safe(filename):
    try:
        with open(filename, "r") as file:
            return file.read()
    except FileNotFoundError:
        return f"Error: File '{filename}' not found."
    finally:
        print("File operation attempt complete.")


# Testing
print(read_file_safe("python_notes.txt"))  # should work
print(read_file_safe("ghost_file.txt"))    # should fail
```

## Step 3 — API Request with Error Handling

This step makes an API request and handles different scenarios like successful response, product not found (404), connection issues, and timeout errors.



```python
# Step 3: API calls with exception handling
import requests

try:
    # Make API call
    response = requests.get("https://dummyjson.com/products/1", timeout=5)

    # Check HTTP response
    if response.status_code == 200:
        data = response.json()
        print("Product Found:")
        print(f"{data['title']} | Price: ${data['price']}")  # PASS case

    elif response.status_code == 404:
        print("Product not found.")  # 404 case

    else:
        print("Unexpected status:", response.status_code)

# Handle connection error
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

# Handle timeout
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

# Handle any other error
except Exception as e:
    print("Error:", e)
```

## Part D
## Step 4 — Input Validation Loop

This step takes user input for product ID, validates it, and fetches product details. It keeps asking until the user enters a valid input or chooses to exit.


```python
# Step 4: Input validation loop
import requests

while True:
    user_input = input("\nEnter product ID (1-100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting...")
        break

    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    if product_id < 1 or product_id > 100:
        print("Please enter ID between 1 and 100.")
        continue

    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url)

        if response.status_code == 404:
            print("Product not found.")
        else:
            product = response.json()
            print(f"{product['title']} | Price: ${product['price']}")

    except Exception as e:
        print("Error:", e)
```

## Task 4 — Logging to File

## Task 4 — Logging Errors

This step logs different types of errors into a file with a timestamp. It helps track issues like connection errors, timeouts, and invalid API responses.

```python
# Import datetime
from datetime import datetime

# Function to log errors
def log_error(message):
    with open("error_log.txt", "a") as file:  # append mode
        time = datetime.now()
        file.write(f"[{time}] ERROR: {message}\n")

```

```python
import requests

# 1. Trigger ConnectionError
try:
    requests.get("https://this-url-does-not-exist-xyz.com")
except requests.exceptions.ConnectionError as e:
    log_error(f"ConnectionError - {e}")


# 2. Trigger 404 error (manual check)
response = requests.get("https://dummyjson.com/products/9999")

if response.status_code != 200:
    log_error(f"HTTP Error {response.status_code} - Product not found")


# 3. Trigger Timeout
try:
    requests.get("https://dummyjson.com/products", timeout=0.001)
except requests.exceptions.Timeout as e:
    log_error(f"Timeout - {e}")
```

```python
# Read and display log file
with open("error_log.txt", "r") as file:
    for i, line in enumerate(file, start=1):
        print(f"{i}. {line.strip()}")

```
